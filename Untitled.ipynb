```python
url = "https://raw.githubusercontent.com/zhenliangma/Applied-AI-in-Transportation/main/ProjectAssignmentData/Dataset-PT.csv"
df = pd.read_csv(url,header=1)
df.head(5)
# df.size 545104 rows
```

```python
for i in range(1, 29):

    desired_stop_sequence = i  # Replace with your specific stop sequence number

    # Create a boolean mask for rows with the desired stop sequence
    mask = df['stop_sequence'] == desired_stop_sequence

    # Filter the DataFrame to keep only the rows with the desired stop sequence
    df_stop = df[mask]

    df_stop.shape
    df_stop_weekday = df_stop[df_stop['day_of_week'] == 'weekday']
    df_stop_weekend = df_stop[df_stop['day_of_week'] == 'weekend']
    from sklearn.model_selection import cross_val_score

    subset_M1_RG = df_stop_weekday[['traffic_condition', 'recurrent_delay']].copy()

    subset_M1_RG['traffic_condition'] = subset_M1_RG['traffic_condition'].astype(int)
    subset_M1_RG['recurrent_delay'] = subset_M1_RG['recurrent_delay'].astype(int)

    df_stop_weekday[['traffic_condition', 'recurrent_delay']] = subset_M1_RG

    x = df_stop_weekday.drop(['arrival_delay', 'day_of_week', 'time_of_day',"weather" ,"temperature"], axis=1)
    y = df_stop_weekday['arrival_delay']

    # Neural Network Model
    Model_weekday_RG = LinearRegression()
    # Try corss validation score first
    cross_val_score(Model_weekday_RG, x, y, cv=5).mean()
```
